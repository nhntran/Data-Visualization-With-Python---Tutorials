###  Load and Display the Dataset

- Using the MNIST Fashion dataset from tf.keras

```python
#!pip install tensorflow
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
print(tf.__version__)
```

```python
### Load the data set
mnist = tf.keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# number of class in the y_train
from collections import Counter
counter = Counter(y_train)
counter
```

<a id='basic'></a>
## 1. BASIC VISUALIZATION 

```python
# plt.gray() set all the images to gray scale
# Should not use plt.gray() which is applied for all other images after that 
# reset to default color:
# plt.set_cmap('viridis')
```

```python
### Show the image
plt.imshow(X_train[0])
plt.show()
```

```python
### Hide axis
plt.imshow(X_train[0])
plt.axis('off')
plt.show()
```

```python
### display different color
plt.subplot(1,2,1)
plt.imshow(X_train[0], cmap = "gray")
plt.subplot(1,2,2)
plt.imshow(X_train[0], cmap = "inferno")
plt.show()
```

<a id='grid'></a>
## 2. GRID VISUALIZATION 

```python
### Get representation of each class
label_index = []
for i in range(len(counter)):
    for j in range(len(y_train)):
        if y_train[j] == i:
            label_index.append(j)
            break
            
### Show representation for each class
plt.figure(figsize = (12,12))
for i in range(len(label_index)):
    plt.subplot(4, 3, i+1)
    plt.imshow(X_train[i])
    plt.title("Class {}".format(i))
plt.show()
```
