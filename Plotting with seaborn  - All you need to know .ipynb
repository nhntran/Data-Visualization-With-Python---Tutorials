## USING BASIC STRUCTURE OF MATPLOTLIB IN SEABORN

- The base of a visualization is the **FIGURE** object.
- Each FIGURE has 1 or more **AXES** objects.
- Each AXES has other elements.
- The right way to create plot so that we can take more control over the layout: define Figure and Axes object
- Seaborn has `ax` parameter to specify upon which Axes a plot will be drawn.

```
    fig = plt.figure() # create the Figure object
    ax = fig.add_axes([.125, .125, .775, .755]) # create Axes object
    sns.countplot(data = df, x = 'num_var', ax = ax) # create plot
```

The list agurment: [.125, .125, .775, .755] - the first 2 elements: the position of the lower-left hand corner (in this case one quarter of the way from the lower-left corner of the Figure) - the last two elements specifying the Axes width and height, respectively.

#### 1.2.  COUNT PLOTS FOR RELATIVE FREQUENCY (PERCENTAGE)

No option. Need to do that manually from the absolute value: just relabel the counts axis in terms of proportions.

````python
# COUNTPLOT - VERTICAL
### Calculate the value for tick marks
cat_order = df_pokemon['generation_id'].value_counts().index
total = df_pokemon.shape[0]
max_count = df_pokemon.generation_id.value_counts().max()
max_y = max_count/total

### Generate tick mark locations and names
tick_val = np.arange(0, max_y, 0.05)
tick_name = ['{:0.2f}'.format(v) for v in tick_val]

### Set color for chart
color = sns.color_palette()[2]

### Plot
sns.countplot(data = df_pokemon, x = 'generation_id', color = color, order = cat_order)
plt.yticks(tick_val * total, tick_name)
plt.ylabel('Proportion');
````

````python
# COUNTPLOT - HORIZONTAL
### Calculate the value for tick marks
cat_order = df_pokemon['generation_id'].value_counts().index
total = df_pokemon.shape[0]
max_count = df_pokemon.generation_id.value_counts().max()
max_x = max_count/total

### Generate tick mark locations and names
tick_val = np.arange(0, max_x, 0.05)
tick_name = ['{:0.2f}'.format(v) for v in tick_val]

### Set color for chart
color = sns.color_palette()[2]

### Plot
sns.countplot(data = df_pokemon, y = 'generation_id', color = color, order = cat_order)
plt.xticks(tick_val * total, tick_name)
plt.xlabel('Proportion');
````

### 4. KERNEL DENSITY ESTIMATION

- Estimating the probability density function of a variable. 
- Usage: If there are relatively few data points available, KDE provides a smooth estimate of the overall distribution of data. These ideas may not be so easily conveyed through histograms, in which the large discreteness of jumps may end up misleading.
- The default settings use a normal-distribution kernel, but most software that can produce KDE plots also include other kernel function options.
- Parameter need to be justified:bandwidth parameter (specifies how wide the density lumps are). Similar to bin width for histograms, we need to choose a bandwidth size that best shows the signal in the data. A too-small bandwidth can make the data look noisier than it really is, and a too-large bandwidth can smooth out useful features that we could use to make inferences about the data.
- Available in the seaborn library.

````python
### Example of KDE plot
# data: only 4 points and need to have smooth estimate of the overall distribution
data = [0.0, 3.0, 3.1, 3.5, 4, 4.5, 8.0, 8.0, 9.0, 10.0, 9.2, 8.5]
plt.figure(figsize = [12, 5])

# plot 1: default distplot
plt.subplot(1,4,1)
sns.distplot(data)

# plot 2: showing kde lumps with the default settings
plt.subplot(1, 4, 2)
sns.distplot(data, hist = False, rug = True, rug_kws = {'color' : 'r'})

# plot 3: kde with narrow bandwidth to show individual probability lumps
plt.subplot(1, 4, 3)
sns.distplot(data, hist = False, rug = True, rug_kws = {'color' : 'r'},
            kde_kws = {'bw' : 1})

# plot 4: choosing a different, triangular kernel function (lump shape)
plt.subplot(1, 4, 4)
sns.distplot(data, hist = False, rug = True, rug_kws = {'color' : 'r'},
            kde_kws = {'bw' : 1.5, 'kernel' : 'tri'});
````

### 5. REGPLOT (SCATTERPLOT WITH REGRESSION FUNCTION)

- By default, the regression function is linear, and includes a shaded confidence region for the regression estimate. 

#### 5.1. SIMPLE REGPLOT

````python
### SIMPLE REGPLOT

sns.regplot(data = df_pokemon, x = 'weight', y = 'height', color = 'g')

plt.xlabel('weight')
plt.ylabel('height');
plt.xlim(-100, df_pokemon.weight.max() + 100);
````

````python
### REGPLOT WITH LOG SCALE TRANSFORMATION

def log_transform(x, inverse = False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10,x)
    
sns.regplot(data = df_pokemon, x = df_pokemon['weight'].apply(log_transform), y = 'height')

tick_locs = [0.1, 0.2, 0.5, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
plt.xticks(log_transform(tick_locs), tick_locs)
plt.xlabel('weight')
plt.ylabel('height');
````

#### 5.2. REGPLOT - TRANSPARENCY

````python
### REGPLOT WITH ADDING TRANSPARENCY
# Using scatter_kws = {'alpha': 0.4}

sns.regplot(data = df_pokemon, x = 'weight', y = 'height',
           scatter_kws = {'alpha': 0.4})

plt.xlabel('weight')
plt.ylabel('height');
plt.xlim(-100, df_pokemon.weight.max() + 100);
````

#### 5.3. REGPLOT  - 3 VARIABLES - ENCODING VIA SIZE

- Point size is a good encoding for 3 numeric variables.

- If the values in your third numeric variable include **negative values**, then you might want to choose a color encoding instead.

````python
### REGPLOT WITH ADDING TRANSPARENCY
# Using scatter_kws = {'alpha': 0.4}

sns.regplot(data = df_pokemon, x = 'weight', y = 'height',
           scatter_kws = {'alpha': 0.4, 's': df_pokemon['speed']})

plt.xlabel('weight')
plt.ylabel('height');
plt.xlim(-100, df_pokemon.weight.max() + 100);

# Based on df_pokemon.speed.describe(), using mean and .25 and .75 quantile to define sizes
sizes = [5, 65, 85]
base_color = sns.color_palette()[0]
legend_obj = []
for s in sizes:
    legend_obj.append(plt.scatter([], [], s = s, color = base_color))
plt.legend(legend_obj, sizes);
````

### 9. FACETING

- Help to handle plots of two or more variables
- The data is divided into disjoint subsets, most often by different levels of a categorical variable. For each of these subsets of the data, the same plot type is rendered on other variables. 
- built-in function is available in seaborn as `sns.FacetGrid()`

#### 9. 1. FACETING - SIMPLE

````python
### SIMPLE FACETING
### Create an instance of the FacetGrid object, specify the feature of each plot
g = sns.FacetGrid(data = df_pokemon, col = 'generation_id',
                 col_wrap = 4) # number of rows in facets, default: a single long row 

#### Specify the plot type and variable(s) that will be plotted in each subset
# Create the same bin_edges for all subplots 
# otherwise, each plot may have a different bin size
bin_edges = np.arange(df_pokemon.attack.min() - 10, df_pokemon.attack.max() + 10, 10)
# Create map
g.map(plt.hist, 'attack', 
        bins = bin_edges # set bin size to be the same across subplot
        );
g.set_titles('{col_name}');
````

#### 9. 2. FACETING - SORTING BY GROUP

````python
### FACETING - MORE CUSTOMIZED
# Organize the subplot in a specific order (for example: mean value from big to small)
group_means = df_pokemon.groupby(['generation_id']).mean()
group_order = group_means.sort_values(['attack'], ascending = False).index


### Create an instance of the FacetGrid object, specify the feature of each plot
g = sns.FacetGrid(data = df_pokemon, col = 'generation_id',
                 col_wrap = 4, # number of rows in facets, default: a single long row
                 height = 2, # setting each facet height to 2 inches
                  col_order = group_order) # sorting the facets by group

#### Specify the plot type and variable(s) that will be plotted in each subset
# Create the same bin_edges for all subplots 
# otherwise, each plot may have a different bin size
bin_edges = np.arange(df_pokemon.attack.min() - 10, df_pokemon.attack.max() + 10, 10)
# Create map
g.map(plt.hist, 'attack', 
        bins = bin_edges # set bin size to be the same across subplot
        );

#changing the titles of each facet to just the categorical level name 
# {col_name} template variable.
g.set_titles('{col_name}');
````

#### 9.3. FACETING FOR SCATTERPLOT - APPLYING LOG SCALE ON FACETING

````python
### Process data to combine type_1 and type_2
type_cols = ['type_1','type_2']
non_type_cols = df_pokemon.columns.difference(type_cols)
pkmn_types = df_pokemon.melt(id_vars = non_type_cols, value_vars = type_cols, 
                          var_name = 'type_level', value_name = 'type').dropna()
### Select to analyze only 'dragon' and 'fairy'
pkm2types = pkmn_types.query('type in ["dragon", "fairy"]')

# log scale transform
def log_transform(x, inverse = False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10,x)
    
### Log transform the weight
pkm2types['log_weight'] = pkm2types['weight']
pkm2types['log_weight'] = pkm2types['log_weight'].apply(log_transform);

## Set figure size
sns.set(rc={'figure.figsize':(20, 10)})

## Create FACETGRID
g = sns.FacetGrid(data = pkm2types, hue = 'type', height=5)
g = g.map(plt.scatter, 'log_weight', 'height')

### Aply real value on logscale value for easy intepreting
tick_locs = [0.1, 0.2, 0.5, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
plt.xticks(log_transform(tick_locs), tick_locs);
````

#### 10.1. FACETING WITH PLT.HIST

````python
### SIMPLE HISTOGRAM LINE PLOT

# Variable to be plotted
num_var = 'attack'
cat_var = 'generation_id'
# calculate bin
bin_edges = np.arange(0, df_pokemon[num_var].max()+10, 10)

# Create plot
g = sns.FacetGrid(data = df_pokemon, hue = cat_var, height = 5)
g.map(plt.hist, num_var, bins = bin_edges, histtype = 'step')
g.add_legend();
````

#### 10.2. FACETING WITH CUSTOMIZED ERRORBAR

````python
### FACETING WITH CUSTOMIZED ERRORBAR

# Variable to be plotted
num_var = 'attack'
cat_var = 'generation_id'

def freq_poly(x, bins = 10, **kwargs):
    """ Custom frequency polygon / line plot code. """
    #**kwargs: allow additional keyword arguments to be set for the errorbar function.
    # set bin edges if none or int specified
    if type(bins) == int:
        bins = np.linspace(x.min(), x.max(), bins+1)
    bin_centers = (bin_edges[1:] + bin_edges[:-1]) / 2

    # compute counts
    data_bins = pd.cut(x, bins, right = False,
                       include_lowest = True)
    counts = x.groupby(data_bins).count()

    # create plot
    plt.errorbar(x = bin_centers, y = counts, **kwargs)

# calculate bin
bin_edges = np.arange(0, df_pokemon[num_var].max()+10, 10)

# Create plot
g = sns.FacetGrid(data = df_pokemon, hue = cat_var, height = 5)
g.map(freq_poly, num_var, bins = bin_edges)
g.add_legend();
````

### 2. CUSTOMIZATION WITH PAIRGRID

- sns.PairGrid **does not** automatically fill in the plots of every available features (attributes) in the dataframe.

- The real benefits of using the PairGrid: when we want to create custom functions to map different information onto the plot. 

````python
# Create an instance of the PairGrid class.
grid = sns.PairGrid(data= df_wine,
                    vars = ['residual sugar', 'citric acid', 
                    'alcohol'], height = 4)
# Map a scatter plot to the upper triangle
grid = grid.map_upper(plt.scatter, color = 'darkred')
# Map a histogram to the diagonal
# map_diag is slightly different: it takes in a function that accepts a SINGLE array
grid = grid.map_diag(plt.hist, bins = 10, color = 'darkred', 
                     edgecolor = 'k')
# Map a density plot to the lower triangle
grid = grid.map_lower(sns.kdeplot, cmap = 'Reds')
````

````python
## Create custom functions to map different information onto the plot. 
# An example:
# Function to calculate correlation coefficient between two arrays
def corr(x, y, **kwargs):
    
    # Calculate the value
    coef = np.corrcoef(x, y)[0][1]
    # Make the label
    label = r'$\rho$ = ' + str(round(coef, 2))
    
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy = (0.2, 0.95), size = 20, xycoords = ax.transAxes)
    
# Create a pair grid instance
grid = sns.PairGrid(data= df_wine,
                    vars = ['residual sugar','citric acid','alcohol'], height = 4)

# Map the plots to the locations
grid = grid.map_upper(plt.scatter, color = 'darkred')
grid = grid.map_upper(corr)
grid = grid.map_lower(sns.kdeplot, cmap = 'Reds')
grid = grid.map_diag(plt.hist, bins = 10, edgecolor =  'k', color = 'darkred');
````

````python
# Function to show the summary statistics
def show_statistics(x, **kwargs):
    
    # Make the label
    label = 'Summary Statistics:\n\n'
    # mean
    label += 'Mean: ' + str(x.mean()) + '\n'
    label += 'Max: ' + str(x.max()) + '\n'
    label += 'Min: ' + str(x.min()) + '\n'
    
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy = (0.2, 0.3), size = 20, xycoords = ax.transAxes)
    
# Create a pair grid instance
grid = sns.PairGrid(data= df_wine,
                    vars = ['residual sugar','citric acid','alcohol'], height = 4)

# Map the plots to the locations
grid = grid.map_upper(plt.scatter, color = 'darkred')
grid = grid.map_lower(sns.kdeplot, cmap = 'Reds')
grid = grid.map_diag(show_statistics);
````
