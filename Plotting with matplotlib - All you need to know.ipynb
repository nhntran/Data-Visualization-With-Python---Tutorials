## A. IMPORTANT CONCEPT ABOUT CHART

### 1. UNIVARIATE VISUALIZATIONS (SINGLE VARIABLES)

- For qualitative variables: bar charts
- For quantitative variables: histograms

Note 1: For nominal data (pure labels without inherent order), it is better to arrange the bars in order of frequency (count) with the most common on the left. Do not do that for Ordinal data (labels with an intrinsic order or ranking).

#### 1. 1. BAR CHARTS

- Using for qualitative variables.
- Note: For nominal data (pure labels without inherent order), it is better to arrange the bars in order of frequency (count) with the most common on the left. Do not do that for Ordinal data (labels with an intrinsic order or ranking).
- Use horizontal chart when label names are long or have a lot of classes (bars)

#### 1. 2. HISTOGRAMS

- Used to plot the distribution of a numeric variable. 
- Values are grouped into continuous bins, and one bar for each bin is plotted depicting the number.
- Parameter need to be justified: bins


- **KERNEL DENSITY ESTIMATION**

    - Estimating the probability density function of a variable. 
    - Usage: If there are relatively few data points available, KDE provides a smooth estimate of the overall distribution of data. These ideas may not be so easily conveyed through histograms, in which the large discreteness of jumps may end up misleading.
    - The default settings use a normal-distribution kernel, but most software that can produce KDE plots also include other kernel function options.
    - Parameter need to be justified:bandwidth parameter (specifies how wide the density lumps are). Similar to bin width for histograms, we need to choose a bandwidth size that best shows the signal in the data. A too-small bandwidth can make the data look noisier than it really is, and a too-large bandwidth can smooth out useful features that we could use to make inferences about the data.
    - Available in the seaborn library.
    
#### 1.3. PIE CHARTS

- A common univariate plot type that is used to depict relative frequencies for levels of a categorical variable.
- If > 5 categories, or categories that have small proportional representation => grouping them together so that fewer wedges are plotted, or use an 'Other' category to handle them.
- A bar chart is a safer choice in general. The bar heights are more precisely interpreted than areas or angles, and a bar chart can be displayed more compactly than a pie chart. 
    
#### 1. 4. WAFFLE PLOTS (SQUARE PIE CHART)

- Use a square (10x10 grid) to represent the whole. Each small square in the grid represents one percent of the data
- No built-in function, utilized plt.bar().
- Need to caculate the percentage blocks.
- The blocks are drawn from left to right, bottom to top, using the ones and tens digits for numbers from 0 to 99 to specify the x- and y- positions, respectively. 
- Can use buil-in waffle chart in the package PyWaffle (https://github.com/gyli/PyWaffle)


#### 1. 5. QUANTILE-QUANTILE PLOT (Q-Q PLOT)

- To see how closely your numeric data follows some hypothetical distribution
- Make a visual comparison between your data and your reference distribution, for example, compare the real data and a hypothetical normal distribution using the sample statistics.

### 2. BIVARIATE VISUALIZATIONS (2 VARIABLES)

- For quantitative + quantitative variables: scatterplots
- For quantitative + qualitative variables: violin plots
- For qualitative + qualitative variables: cluster bar charts (bar chart and heat map)


#### 2. 1. SCATTERPLOTS

- Using for 2 quantitative variables
- Used to check the Pearson correlation coefficient in data (statistic quantifying the strength of linear correllation betweeen 2 numeric variables).

**OVERPLOTTING**: 
    - The high amount of overlap in points makes it difficult to see the actual relationship between the plotted variables.
    - Reason: Due to a very large number of points to plot or our numeric variables are discrete-valued.

**SOLUTION FOR OVERPLOTTING**: 
    - Transparency: Available in both matplotlib and seaborn with `alpha` parameter: 0 - fully transparent, not visible, 1 - fully opaque.
    - Jitter: Only have built-in function in seaborn. x- and y- jitter can be added independently, and won't affect the fit of any regression function. Each point to be plotted in a uniform Â±0.2 range of their true values

#### 2. 2. HEAT MAPS

- Using for 2 quantitative variables
- A 2-d version of the histogram that can be used as an alternative to a scatterplot

#### 2. 3. VIOLIN PLOTS

- Using for 1 quantitative and 1 qualitative variable
- For each level of the categorical variable, a distribution of the values on the numeric variable is plotted. The distribution is plotted as a kernel density estimate, something like a smoothed histogram.
- May be a better option for exploration

#### 2. 4. BOX PLOTS

- Using for 1 quantitative and 1 qualitative variable
- Compared to the violin plot, the box plot leans more on summarization of the data, primarily just reporting a set of descriptive statistics for the numeric values on each categorical level.

**BOX PLOT CHARACTERISTICS:**

- Central line: median of the distribution
- Top and bottom: the third and first quartiles of the data, respectively
- The height of the box: interquartile range (IQR). 
- The whiskers: the range from the first or third quartiles to the minimum or maximum value in the distribution.
- Amaximum range is set on whisker length; by default this is 1.5 times the IQR. 
- Points below/above the whisker: individual outlier points that are more than 1.5 times the IQR below the first quartile.

#### 2. 5. CLUSTERED BAR CHARTS

- Using for 2 qualitative variables
- It's easy to plot in seaborn (sns.countplot, sns.heatmap)

#### 2. 6. BAR CHARTS DISPLAYING MEAN

- Using for 1 quantitative and 1 qualitative variable
- If the numeric variable is binary in nature, taking values only of 0 or 1, then a box plot or violin plot will not be informative, leaving the adapted bar chart as the best choice for displaying the data.
- Mean of data is the mean of binary value, in a range [0,1] => Can see easily in a bar chart.

#### 2. 7. LINE PLOTS - "ERRORBAR"

- Using for 1 quantitative and 1 qualitative variable

- Using when need to emphasize relative change and trends across x-values
- If the x-variable represents time, then a line plot of the data is frequently known as a time series plot (sns.tsplot built-in plot)

## B. THE BASIC STRUCTURE OF VISUALIZATIONS USING MATPLOTLIB

- The base of a visualization is the **FIGURE** object.
- Each FIGURE has 1 or more **AXES** objects.
- Each AXES has other elements.
- The right way to create plot so that we can take more control over the layout: define Figure and Axes object

```
    fig = plt.figure() # create the Figure object
    ax = fig.add_axes([.125, .125, .775, .755]) # create Axes object
    ax.hist(data = df, x = 'num_var') # create plot
```
The list agurment: [.125, .125, .775, .755] - the first 2 elements: the position of the lower-left hand corner (in this case one quarter of the way from the lower-left corner of the Figure) - the last two elements specifying the Axes width and height, respectively.

- Seaborn has `ax` parameter to specify upon which Axes a plot will be drawn.

```
    fig = plt.figure() # create the Figure object
    ax = fig.add_axes([.125, .125, .775, .755]) # create Axes object
    sns.countplot(data = df, x = 'num_var', ax = ax) # create plot
```

### 3. WAFFLE PLOTS (SQUARE PIE CHART)

#### 3.1. WAFFLE PLOTS - PERCENTAGE

````python
### FUNCTION TO CALCULATE THE PERCENTAGE
def percentage_blocks(df, var):
    """
    Take as input a dataframe and variable, and return a Pandas series with
    approximate percentage values for filling out a waffle plot.
    """
    # compute base quotas
    percentages = 100 * df[var].value_counts() / df.shape[0]
    counts = np.floor(percentages).astype(int) # integer part = minimum quota
    decimal = (percentages - counts).sort_values(ascending = False)

    # add in additional counts to reach 100
    rem = 100 - counts.sum()
    for cat in decimal.index[:rem]:
        counts[cat] += 1

    return counts
````

````python
### MAKE WAFFLE PLOTS

waffle_counts = percentage_blocks(df_pokemon, 'generation_id')
prev_count = 0
# for each category,
for cat in range(waffle_counts.shape[0]):
    # get the block indices
    blocks = np.arange(prev_count, prev_count + waffle_counts.values[cat])
    
    # and put a block at each index's location
    x = blocks % 10 # use mod operation to get ones digit
    y = blocks // 10 # use floor division to get tens digit
    plt.bar(x = x, height = 0.8, width = 0.8, bottom = y)
    prev_count += waffle_counts.values[cat]

# aesthetic wrangling
plt.legend(waffle_counts.index, bbox_to_anchor = (1, 0.5), loc = 6)
plt.axis('off')
plt.axis('square');
````

#### 3.2. WAFFLE PLOTS - ABSOLUTE VALUE

````python
### FUNCTION TO CALCULATE THE PERCENTAGE
data = df_pokemon['generation_id']

n = data.value_counts().max()//15 # each box represents n full counts

waffle_counts = (df_pokemon['generation_id'].value_counts() / n).astype(int)

prev_count = 0
# for each category,
for cat in range(waffle_counts.shape[0]):
    # get the block indices
    blocks = np.arange(prev_count, prev_count + waffle_counts.values[cat])
    # and put a block at each index's location
    x = blocks % 10
    y = blocks // 10
    plt.bar(y, 0.8, 0.8, x)
    prev_count += waffle_counts.values[cat]

# box size legend
x_box = 10
y_box = 1.6

plt.bar(10, 0.8, 0.8, 1.8, color = 'white', edgecolor = 'black', lw = 2)
plt.text(x_box + 0.6, y_box + 0.4, f"= {n} data points", va = 'center')

# aesthetic wrangling
plt.legend(waffle_counts.index, bbox_to_anchor = (0.9, 0.5), loc = 6)
plt.axis('off')
plt.axis('square');
````

### 6. LINE PLOTS - "ERRORBAR"

- Need to process and summarize the data before plot
- If 2 variables are both numeric: set a number of bins into which the data will be grouped.

#### 6.1. SIMPLE ERRORBAR PLOT -  2 QUANTITATIVE VARIABLES

````python
### SIMPLE ERRORBAR PLOT 

### Process the data
# Variables to be plotted
var1 = df_pokemon['attack']
var2 = df_pokemon['speed']

# set bin edges, compute centers
bin_size = 10
xbin_edges = np.arange(0, var1.max()+bin_size, bin_size) # bin edges
xbin_centers = (xbin_edges + bin_size/2)[:-1] # for x location

# compute statistics in each bin
data_xbins = pd.cut(var1, xbin_edges, right = False, include_lowest = True)
y_means = var2.groupby(data_xbins).mean() # mean
y_sems = var2.groupby(data_xbins).sem() # standard error of the mean

### Plot the summarized data
plt.errorbar(x = xbin_centers, y = y_means, yerr = y_sems)
plt.xlabel('attack')
plt.ylabel('speed')
````

#### 6.2. ERRORBAR PLOT WITH ROLLING - 2 QUANTITATIVE VARIABLES

#### 2. XSCALE FUNCTION

- Using xscale funtion, don't change the value of the data, only change how they are displayed
- Note: Even though data is on logscale, bins are still LINEARLY spaced + labels are not good
=> Need to change bin_edges and tick value
- Setting ticks in cycles of 1-3-10 or 1-2-5-10 are very useful for base-10 log transforms. Hard code tick_locs base on summary statistics (looking at min, max). Example of ticks:
```
    tick_locs = [0.3, 1, 3, 10, 30, 100, 300, 1000]
    tick_locs = tick_locs = [0.2, 0.5, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
```

````python
### SCALES AND TRANSFORMATION
### USING XSCALE, NOT CHANGE THE VALUE, ONLY CHANGE HOW THEY'RE DISPLAYED

plt.figure(figsize = [10, 5])

### LEFT HISTOGRAM: data plotted using xscale function
plt.subplot(1, 2, 1)
bin_edges = np.arange(0, df_pokemon.weight.max() + 10, 10)
plt.hist(df_pokemon.weight, bins = bin_edges)
plt.xscale('log')

### RIGHT HISTOGRAM: data plotted using xscale function + modified bin_edges and tick value
plt.subplot(1, 2, 2)
# Changing bins
bin_edges = 10 ** np.arange(np.log10(df_pokemon.weight.min())- 0.1, np.log10(df_pokemon.weight.max()) + 0.1, 0.1)
# Plot normal data using xscale function
plt.hist(df_pokemon.weight, bins = bin_edges)
plt.xscale('log')

# Hard code tick_locs base on summary statistics (looking at min, max and decide what range to put)
tick_locs = [0.2, 0.5, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
plt.xticks(tick_locs, tick_locs)
# change the stick label
plt.xticks(rotation = 90);
````

#### 3. CUSTOMIZED FUNCTION

- When need to use a different transformation that's not available in xscale, for example: square root function.
=> Need to writing a function that applies both the transformation and its inverse. 
- The inverse will be useful in cases where we specify values in their transformed units and need to get the natural units back.

````python
### Example of customized transformation using square root

def sqrt_trans(x, inverse = False):
    """ transformation helper function """
    if not inverse:
        return np.sqrt(x)
    else:
        return x ** 2
    
data = df_pokemon.weight
bin_edges = np.arange(0, sqrt_trans(data.max())+1, 1)
plt.hist(data.apply(sqrt_trans), bins = bin_edges) # used apply because data is panda.Series
tick_locs = np.arange(0, sqrt_trans(data.max())+10, 10)
plt.xticks(tick_locs, sqrt_trans(tick_locs, inverse = True).astype(int));
````

### 2. CREATE SUBPLOT - SHORTCUT

#### Don't assign Axes objects as they're created#
- Even if we don't assign any variable to return the function's output, Python will still implicitly know that further plotting calls that need a Figure will refer to that Figure as the active one.
- Code:
```
    plt.figure(figsize = [10, 5]) # larger figure size for subplots
    plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
    plt.hist(data = df, x = 'Age');

    plt.subplot(1, 2, 2)# 1 row, 2 cols, subplot 2
    plt.hist(data = df, x = 'Speed');
```
- `plt.subplot(1, 2, 1)` will create a new Axes in our Figure. The argument will help divide the figure into 1 row and 2 columns, and the third argument says to create a new Axes in the first slot.

- Slots are numbered from left to right in rows from top to bottom. Note in particular that the index numbers start at 1 (rather than the usual Python indexing starting from 0). 

- If you don't assign Axes objects as they're created, you can retrieve the current Axes using:
```
    ax = plt.gca()
    # or
    axes = fig.get_axes()
```

````python
### Create the Figure object.
plt.figure(figsize = [10, 5]) # larger figure size for subplots

### Creates a new Axes in the current Figure
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
bin_edges = np.arange(0, df_pokemon['speed'].max()+1, 10)
plt.hist(data = df_pokemon, x = 'speed', bins = bin_edges);

# example of somewhat too-small bin size
plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
bin_edges = np.arange(0, df_pokemon['speed'].max()+1, 3)
plt.hist(data = df_pokemon, x = 'speed', bins = bin_edges);
````
